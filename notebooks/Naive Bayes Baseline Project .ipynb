```python
df_raw = pd.read_csv('gcj2008.csv')
df_raw.head()
```

```python
df=df_raw.dropna()
```

```python
df=df[:10000]
df
```

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

target_encoder = LabelEncoder().fit(df['username']) 
y = target_encoder.transform(df['username'])
y
```

```python
import numpy as np

from sklearn.model_selection import cross_validate
from sklearn.pipeline import make_pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import recall_score

# Feature/Target
X = df["flines"]

# Pipeline vectorizer + Naive Bayes
pipeline_naive_bayes = make_pipeline(
    TfidfVectorizer(), 
    MultinomialNB()
)

# Cross-validation
cv_results = cross_validate(pipeline_naive_bayes, X, y, cv = 5, error_score='raise')
print(cv_results)
average_recall = cv_results["test_score"].mean()
np.round(average_recall,2)

#scoring = ["recall"]
```
